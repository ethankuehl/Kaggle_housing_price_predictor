```python
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
# null prediction is the mean of y
df = pd.read_csv('datasets/train.csv')
```

```python
pd.set_option('display.max_rows', 999)
pd.set_option('display.max_columns', 999)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
X = df[['Lot Area', 'Overall Cond', 'Gr Liv Area', 'Bedroom AbvGr', 'Kitchen AbvGr']]
y = df['SalePrice']
X.dtypes
```

```python
lr = LinearRegression()
lr.fit(X,y)
```

```python
lr.score(X,y)
```

```python
df_test = pd.read_csv('datasets/test.csv')
```

```python
X_test = df_test[['Lot Area', 'Overall Cond', 'Gr Liv Area', 'Bedroom AbvGr', 'Kitchen AbvGr']]
```

```python
test_preds = lr.predict(X_test)
```

```python
test_saleprice = pd.DataFrame(df_test['Id'])
test_saleprice['SalePrice'] = test_preds
```

```python
test_saleprice
```

```python
test_saleprice.to_csv('output/prediction_1.csv', index=False)
```
